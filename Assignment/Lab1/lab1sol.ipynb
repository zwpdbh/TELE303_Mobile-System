## DFT

We have played a bit with waveforms, but what about the spectrum? We need to use Fourier transform,
which decomposes a signal into a set of sine waves. For a digital signal the Discrete Fourier Transform
extracts the frequency components on discrete intervals $f_k = 2k\pi/N (k=0,1,..., N-1)$ for a signal of length
$N$). The series of the amplitudes over the N frequency levels makes the power spectrum of the signal. FFT
is a fast algorithm for DFT.

FFT can get quite complicated, but fortunately DFT/FFT has been implemented in NumPy. Have a try
with the following:

```python
%matplotlib inline
# get the PI value
PI=np.pi
# specify sampling rate (Hz) and sampling time length (s)
samplrate=1000
nsec=1.0   # 1 second
# get the discrete sampling time sequence
#t=sp.r_[0:nsec:1.0/samplrate]
t=np.arange(0,nsec,1.0/samplrate)
# here's a signal (f=? Hz)
sig=np.sin(2*PI*10*t)
# plot out the waveform
pl.figure(1)
pl.plot(sig)
pl.title('Fig.1. Original sig')

# conduct FFT
fftout=np.fft.fft(sig)
# get the power spectrum
ps=np.abs(fftout)
# the rest is for visualisation
pl.figure(2)
pl.plot(ps)
pl.xlabel('n')
pl.ylabel('Amplitude')
pl.title('Fig.2. After FFT')

##'''
s_res=np.fft.ifft(fftout).real
pl.figure(3)
pl.plot(s_res)
pl.title('Fig.3. After IFFT')
```

**Q6** Explain the spikes observed in the spectrum plot.
- Answer: From the power spectrum we can clearly see a spike at position 10, related to the 10Hz frequency of the original signal. 

There is an interesting mirrored spike, at position 990 though. As there can't be 990Hz frequency components in the signal, it is only a "shadow" of the 10Hz frequency. 

The following is FYI only - you are **not** required to play with any DFT equations. Feel free to skip to the next section.

Mathematically, the central symmetry of the power spectrum is due to the way DFT is defined. Let me explain this using the DFT formula. With DFT, the frequency spectrum is defined on $N$ discrete levels:
$$X(k)=\frac{1}{N}\sum_{n=0}^{N-1}x[n]e^{-jk\frac{2\pi}{N}n},$$
for $n=0,1,...,N-1.

On the mirrored location, we have
$$X(N-k)=\frac{1}{N}\sum_{n=0}^{N-1}x[n]e^{-j(N-k)\frac{2\pi}{N}n}=\frac{1}{N}\sum_{n=0}^{N-1}x[n]e^{-jN\frac{2\pi}{N}n}e^{jk\frac{2\pi}{N}n}=\frac{1}{N}\sum_{n=0}^{N-1}x[n]e^{-j2\pi n}e^{jk\frac{2\pi}{N}n},$$
Since $e^{-j2\pi n}=\cos(2\pi n)-j\sin(2\pi n)=1$, this simplifies to
$$X(N-k)=\frac{1}{N}\sum_{n=0}^{N-1}x[n]e^{jk\frac{2\pi}{N}n}=X^{*}(k),$$
i.e., $X(N-k)$ is the conjugate of $X(k)$, and the norm of both from the abs() should be equal. QED

## The Problem of Dial Tones

Let us make this lab a little more interesting by looking into a real world application. Dual Tone Multi-Frequency (DTMF) is an important technique used in wireline and mobile phones. Suppose we are using a
call centre service and need to key in some numbers (e.g., a menu selection, or a PIN). How does the call
centre automatically make out which key/number has been pressed on your phone?
DTMF uses a simple addition of two frequency components to indicate which key on a telephone touch
pad is being pressed. The frequency-digit relationship is listed below:

| Freq.  | 1209Hz | 1336Hz | 1477 Hz
| :- | :-: | :-: | :-: 
| 697Hz  |   1    |   2    |   3    
| 770Hz  |   4    |   5    |   6   
| 852Hz  |   7    |   8    |   9
| 941Hz  |   *    |   0    |   #

Here the column corresponds to a high frequency, and the row corresponds to the low frequency. For
example, pressing the hash key # would generate a signal as the sum of two sine waves, one in 1477 Hz (high
frequency), and another in 941 Hz (low frequency).

**Task B**. Use the example code and generate the waveform for digit '5' by using two frequencies; 
display the power spectrum. Note you'll need to set an appropriate sampling rate. Carry out FFT on the composite signal and display the spectrum. 


```python
## Code for Task B 
samp_freq=3000
nsec=1.0
t=np.arange(0,nsec,1.0/samp_freq)

# this generates a composite signal for '3' 
pf1=np.sin(2*np.pi*1477*t)
pf2=np.sin(2*np.pi*697*t)
psig=pf1+pf2

pl.figure(1)
# Cut and show the first 100 of the 3000 data points
pl.plot(t[0:100],psig[0:100])

pl.figure(2)
pdft=np.fft.fft(psig)
omega=np.fft.fftfreq(len(psig))
amp=abs(pdft)
pl.plot(amp)   
```

**Comments**: The two spikes, positioned on the corresponding frequencies, are clearly shown (along with two mirrored spikes on the right). Note you must specify a sampling rate that satisfies the Nyquist rule. 
In this case, you should use a sampling rate at least of 3000 Hz because of the much
higher frequency values corresponding to the keys (about 1500Hz).

The receiving end can carry out FFT of the incoming signal and by examining the spectrum whether 
there is a particular pair of frequencies one can tell what digit has been pressed.

The DTMF is actually a smart idea - not only it naturally leads to a square
layout (of the keypad) easy to access (compared with a long line-up of all 12 keys!),
but also it saves the analog circuit components by using fewer frequencies. If
somehow we decided each key should be represented by a different frequency (i.e. plain "frequecy shift keying"), we’d
need 12 different circuits to generate 12 frequencies for the 12 keys. In DTMF we
only need 7 frequencies! Viva engineers.
